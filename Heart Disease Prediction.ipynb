## Heart Disease Prediction Using Logistic Regression Machine

```python
# importing the library 
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
heart_data = pd.read_csv(r"C:\Users\utkar\100 days ML\Heart Disease Prediction\heart_disease_data.csv")
```

```python
# first 5 row of the data
heart_data.head()
```

```python
# Cheching for missing values 
heart_data.isnull().sum()
```

```python
# no. of rows and columns in the dataset
heart_data.shape
```

```python
# getting the information about dataset 
heart_data.info
```

```python
# getting statistical values about data
heart_data.describe()
```

```python
# checking the distribution of target variable 
heart_data['target'].value_counts() 
# 1 for defective heart and 2 for healthy heart
```

```python
# spliting the feature and target
x = heart_data.drop('target', axis = 1)
y = heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
# Spliting the data into training data & test data
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size= 0.2,stratify=y,random_state=2)
```

```python
# number of normal , train , test data 
print(x.shape , X_train.shape,X_test.shape)
```

```python
# model training 
model = LogisticRegression()
```

```python
# training the LogisticRegression model with training data
model.fit(X_train,Y_train)
```

```python
# Model evaluation 
# Accuracy on training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("The accuracy of training data : ",training_data_accuracy)
```

```python
# Accuracy on test data 
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("The accuracy of test data : ",test_data_accuracy)
```

```python
# predictive system 
#input_data = (56,1,0,132,184,0,0,105,1,2.1,1,1,1)
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)
# change the input data to a numpy aaray 
data = np.asarray(input_data)
# reshape the array 
reshape_data = data.reshape(1,-1)
# prediction
prediction = model.predict(reshape_data)
print(prediction)
```
